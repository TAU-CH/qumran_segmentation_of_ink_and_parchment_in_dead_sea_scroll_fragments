```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Path to the 16-bit grayscale image
first_band_path = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/124_009_001.tif"

# Read the image in 16-bit grayscale
first_band_image = cv2.imread(first_band_path, cv2.IMREAD_UNCHANGED)

# Display the original and gamma-corrected images
plt.imshow(first_band_image, cmap='gray', vmin=0, vmax=65535)
plt.axis('off')  # Turn off axes
plt.show()
```

```python
# Calculate the histogram
hist = cv2.calcHist([first_band_image], [0], None, [65536], [0, 65536])

# Plot the histogram
plt.plot(hist)
plt.xlim([0, 65535])  # Adjust the x-axis limit as needed
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.savefig('raw_first_band_histogram.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Normalize the image to the range [0, 1] for gamma correction
normalized_first_band_image = first_band_image / 65535.0

# Apply gamma correction (gamma = 2.2 is common, adjust as needed)
gamma = 2.2
gamma_corrected = np.power(normalized_first_band_image, 1.0 / gamma)

# Normalize the gamma-corrected image to the range [0, 65535]
gamma_corrected_normalized_first_band = cv2.normalize(gamma_corrected, None, 0, 65535, cv2.NORM_MINMAX)

# Convert back to 16-bit
gamma_corrected_normalized_first_band = np.uint16(gamma_corrected_normalized_first_band)

# Display the original and gamma-corrected images
plt.imshow(gamma_corrected_normalized_first_band, cmap='gray', vmin=0, vmax=65535)
plt.axis('off')  # Turn off axes
plt.show()
```

```python
# Calculate the histogram
hist = cv2.calcHist([gamma_corrected_normalized_first_band], [0], None, [65536], [0, 65536])

# Plot the histogram
plt.plot(hist)
plt.xlim([0, 65535])  # Adjust the x-axis limit as needed
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.savefig('normalized_first_band_histogram.png', dpi=300, bbox_inches='tight')
plt.show()
```
