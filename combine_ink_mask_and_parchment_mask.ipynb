```python
import os
import cv2
import numpy as np

ink_masks_dir = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/ink_masks"
temp_parchment_masks_dir = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/temp_parchment_masks"
parchment_masks_dir = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/parchment_masks"
background_masks_dir = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/background_masks"
colored_ink_parchment_masks_dir = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/colored_ink_parchment_masks"

if not os.path.exists(parchment_masks_dir):
    os.makedirs(parchment_masks_dir)
if not os.path.exists(background_masks_dir):
    os.makedirs(background_masks_dir)
if not os.path.exists(colored_ink_parchment_masks_dir):
    os.makedirs(colored_ink_parchment_masks_dir)

for filename in os.listdir(ink_masks_dir):
    if filename.endswith(".png"):
        ink_mask_path = os.path.join(ink_masks_dir, filename)
        temp_parchment_mask_path = os.path.join(temp_parchment_masks_dir, filename)
        parchment_mask_path = os.path.join(parchment_masks_dir, filename)
        background_mask_path = os.path.join(background_masks_dir, filename)
        colored_ink_parchment_mask_path = os.path.join(colored_ink_parchment_masks_dir, filename)

        ink_mask = cv2.imread(ink_mask_path, cv2.IMREAD_GRAYSCALE)
        temp_parchment_mask = cv2.imread(temp_parchment_mask_path, cv2.IMREAD_GRAYSCALE)

        # OR operation
        parchment_mask = cv2.bitwise_or(ink_mask, temp_parchment_mask)
        cv2.imwrite(parchment_mask_path, parchment_mask)

        # Invert and save background mask
        background_mask = cv2.bitwise_not(parchment_mask)
        cv2.imwrite(background_mask_path, background_mask)


        # Create color image
        colored_mask = np.zeros((ink_mask.shape[0], ink_mask.shape[1], 3), dtype=np.uint8)
        colored_mask[background_mask > 0] = [255, 0, 0]  # Blue for background
        colored_mask[parchment_mask > 0] = [0, 255, 0]  # Green for parchment
        colored_mask[ink_mask > 0] = [0, 0, 255]  # Green for ink

        cv2.imwrite(colored_ink_parchment_mask_path, colored_mask)
```
