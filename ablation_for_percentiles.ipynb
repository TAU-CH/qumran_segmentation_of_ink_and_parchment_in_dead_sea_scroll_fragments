```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import maxflow
# Updated output directories
ink_output_dir = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/msem_results/ablation/ink_masks_with_percentiles/"
parchment_output_dir = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/msem_results/ablation/parchment_masks_with_percentiles/"
input_base = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/qumran_segmentation_dataset/multispectral_images/"


# Ensure output directories exist
os.makedirs(ink_output_dir, exist_ok=True)
os.makedirs(parchment_output_dir, exist_ok=True)

def create_mask(i1, i12, conditions):
    mask = np.ones_like(i1, dtype=bool)
    for cond in conditions:
        mask &= cond
    result = np.zeros_like(i1)
    result[mask] = 255
    return np.uint8(result)

# List of fragments to process
fragments = ['124_001']  # Only process fragment 124_001

# Threshold values for each percentile set
thresholds = {
    6: {'parchment': (156.0, 1230.0), 'ink_band1': (87.0, 163.0), 'ink_band12_band1': (71.0, 310.0)},
    8: {'parchment': (219.0, 1202.0), 'ink_band1': (90.0, 156.0), 'ink_band12_band1': (75.0, 294.0)},
    10: {'parchment': (313.0, 1179.0), 'ink_band1': (92.0, 150.0), 'ink_band12_band1': (78.0, 280.0)},
    12: {'parchment': (403.0, 1159.0), 'ink_band1': (94.0, 146.0), 'ink_band12_band1': (81.0, 267.0)},
    14: {'parchment': (476.0, 1142.0), 'ink_band1': (96.0, 142.0), 'ink_band12_band1': (84.0, 254.0)},
}

# Iterate over each threshold set
for n, values in thresholds.items():
    print(f"Processing fragment: 124_001 with threshold set for n = {n}")

    # Load images for fragment 124_001
    i1_path = os.path.join(input_base, '124_001_001.tif')
    i12_path = os.path.join(input_base, '124_001_012.tif')
    i1 = cv2.imread(i1_path, cv2.IMREAD_UNCHANGED).astype(np.float64)
    i12 = cv2.imread(i12_path, cv2.IMREAD_UNCHANGED).astype(np.float64)

    # Thresholds for current percentile set
    parchment_min, parchment_max = values['parchment']
    ink_band1_min, ink_band1_max = values['ink_band1']
    ink_band12_band1_min, ink_band12_band1_max = values['ink_band12_band1']

    # Update masks with current thresholds
    parchment_mask = create_mask(i1, i12, [
        i12 - i1 > parchment_min
    ])
    ink_mask = create_mask(i1, i12, [
        i12 - i1 > ink_band12_band1_min, i12 - i1 <= ink_band12_band1_max, i1 > ink_band1_min, i1 <= ink_band1_max
    ])
    ink_contour_mask = create_mask(i1, i12, [
        i12 - i1 > 225, i12 - i1 <= 346, i1 > ink_band1_min, i1 <= ink_band1_max
    ])

    # Apply rest of MSEM pipeline
    inverse_parchment_mask = cv2.bitwise_not(parchment_mask)
    parchment_and_ink_contour_mask = cv2.bitwise_or(parchment_mask, ink_contour_mask)
    parchment_and_ink_contour_and_ink_mask = cv2.bitwise_or(parchment_and_ink_contour_mask, ink_mask)
    inverse_parchment_and_ink_contour_and_ink_mask = cv2.bitwise_not(parchment_and_ink_contour_and_ink_mask)

    # Graph-cut segmentation function
    def graph_cut_segmentation(imagetoseg, h, i):
        white_pixels_mask = imagetoseg > 0
        dist_transform_h = cv2.distanceTransform(255 - h, cv2.DIST_L2, 5)
        dist_transform_i = cv2.distanceTransform(255 - i, cv2.DIST_L2, 5)
        data_cost_h = np.where(white_pixels_mask, 1 / (1 + dist_transform_h), np.inf)
        data_cost_i = np.where(white_pixels_mask, 1 / (1 + dist_transform_i), np.inf)

        num_white_pixels = np.sum(white_pixels_mask)
        g = maxflow.Graph[float](num_white_pixels, num_white_pixels * 4)
        node_ids = g.add_nodes(num_white_pixels)

        white_pixel_indices = np.argwhere(white_pixels_mask)
        node_id_map = np.full(white_pixels_mask.shape, -1, dtype=int)
        node_id_map[white_pixels_mask] = node_ids

        for node_id, (y, x) in enumerate(white_pixel_indices):
            g.add_tedge(node_id, data_cost_h[y, x], data_cost_i[y, x])

        for y, x in white_pixel_indices:
            if y + 1 < white_pixels_mask.shape[0] and white_pixels_mask[y + 1, x]:
                g.add_edge(node_id_map[y, x], node_id_map[y + 1, x], 1, 1)
            if x + 1 < white_pixels_mask.shape[1] and white_pixels_mask[y, x + 1]:
                g.add_edge(node_id_map[y, x], node_id_map[y, x + 1], 1, 1)

        g.maxflow()
        result = np.zeros_like(white_pixels_mask, dtype=bool)
        for node_id, (y, x) in enumerate(white_pixel_indices):
            result[y, x] = g.get_segment(node_id)

        return result.astype(np.uint8) * 255

    riceholeback = inverse_parchment_and_ink_contour_and_ink_mask
    clean_ink_contour = graph_cut_segmentation(ink_contour_mask, riceholeback, parchment_mask)
    ink = graph_cut_segmentation(inverse_parchment_mask, parchment_mask, clean_ink_contour)
    # Clean the masks with morphological operations
    def clean_mask(mask):
        kernel = np.ones((5, 5), np.uint8)
        clean_white = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
        inverted_image = cv2.bitwise_not(clean_white)
        clean_black = cv2.morphologyEx(inverted_image, cv2.MORPH_OPEN, kernel)
        return cv2.bitwise_not(clean_black)
    # Clean the masks
    clean_ink = clean_mask(ink)
    inkpar = cv2.bitwise_or(parchment_mask, ink)
    inkpar = cv2.bitwise_or(inkpar, clean_ink)
    clean_inkpar = clean_mask(inkpar)
    background = cv2.bitwise_not(clean_inkpar)

    # Output file paths for this threshold set
    ink_output_path = os.path.join(ink_output_dir, f"124_001_n{n}.png")
    parchment_output_path = os.path.join(parchment_output_dir, f"124_001_n{n}.png")
    cv2.imwrite(ink_output_path, clean_ink)
    cv2.imwrite(parchment_output_path, clean_inkpar)

    # Visualization
    redink_blueback_greenparchment = np.zeros((i1.shape[0], i1.shape[1], 3), dtype=np.uint8)
    redink_blueback_greenparchment[clean_inkpar == 255] = [0, 255, 0]
    redink_blueback_greenparchment[background == 255] = [255, 0, 0]
    redink_blueback_greenparchment[clean_ink == 255] = [0, 0, 255]

    # Display and save visualization
    plt.figure(figsize=(15, 15))
    plt.imshow(redink_blueback_greenparchment[:, :, ::-1])
    plt.axis('off')
    plt.show()
```

```python
import numpy as np
import os
import re
import cv2
import matplotlib.pyplot as plt

# Function to evaluate segmentation metrics
def evaluate_segmentation(ground_truth, prediction):
    ground_truth = ground_truth.flatten()
    prediction = prediction.flatten()
    TP = np.sum((ground_truth == 1) & (prediction == 1))
    FP = np.sum((ground_truth == 0) & (prediction == 1))
    FN = np.sum((ground_truth == 1) & (prediction == 0))
    iou = TP / (TP + FP + FN) if (TP + FP + FN) != 0 else 0
    dice = (2 * TP) / (2 * TP + FP + FN) if (2 * TP + FP + FN) != 0 else 0
    precision = TP / (TP + FP) if (TP + FP) != 0 else 0
    recall = TP / (TP + FN) if (TP + FN) != 0 else 0
    f1_score = (2 * precision * recall) / (precision + recall) if (precision + recall) != 0 else 0
    return {"IoU": iou, "Dice Coefficient": dice, "Precision": precision, "Recall": recall, "F1-Score": f1_score}

# Load specific image from a path
def load_specific_image_from_path(path):
    if os.path.exists(path):
        return cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    return None

# Load images with n value from filenames (only for 124_001.png)
def load_specific_images_with_n(folder, filename_prefix="124_001"):
    images = {}
    for fname in os.listdir(folder):
        if fname.startswith(filename_prefix):
            match = re.search(r'_n(\d+)\.png$', fname)
            if match:
                n = int(match.group(1))
                path = os.path.join(folder, fname)
                images[n] = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    return images

# Evaluate segmentation results for specific image across n values
def evaluate_by_percentile(ground_truth_path, prediction_images_by_n):
    all_metrics = {}
    ground_truth = load_specific_image_from_path(ground_truth_path)

    if ground_truth is None:
        print("No ground truth image found for evaluation.")
        return all_metrics

    _, ground_truth = cv2.threshold(ground_truth, 127, 1, cv2.THRESH_BINARY)

    for n, prediction in prediction_images_by_n.items():
        _, prediction = cv2.threshold(prediction, 127, 1, cv2.THRESH_BINARY)
        result = evaluate_segmentation(ground_truth, prediction)
        all_metrics[n] = result

    return all_metrics

# Define ground truth path
ground_truth_path_ink = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/qumran_segmentation_dataset/ink_masks/124_001.png"
ground_truth_path_parchment = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/qumran_segmentation_dataset/parchment_masks/124_001.png"
ink_output_dir = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/msem_results/ablation/ink_masks_with_percentiles/"
parchment_output_dir = "/content/drive/MyDrive/Colab Notebooks/icdar_ijdar_2025/msem_results/ablation/parchment_masks_with_percentiles/"

# Load predictions by n value (only for 124_001.png)
prediction_images_by_n_ink = load_specific_images_with_n(ink_output_dir)
prediction_images_by_n_parchment = load_specific_images_with_n(parchment_output_dir)

# Evaluate segmentation and plot results (only for 124_001.png)
metrics_by_percentile_ink = evaluate_by_percentile(ground_truth_path_ink, prediction_images_by_n_ink)
metrics_by_percentile_parchment = evaluate_by_percentile(ground_truth_path_parchment, prediction_images_by_n_parchment)

plot_metrics_by_percentile(metrics_by_percentile_ink)
plot_metrics_by_percentile(metrics_by_percentile_parchment)
```
